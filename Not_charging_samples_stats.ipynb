```python
import pandas as pd
import time
import numpy as np
from IPython.display import clear_output
import matplotlib.pyplot as plt
%matplotlib inline

devices_data = pd.read_csv('C:\\Users\\ruben\\Desktop\\Investigation\\updated_dataset_07_3_2018\\devices.csv', index_col='id', parse_dates=['created_at', 'updated_at'])
samples_data = pd.read_csv('C:\\Users\\ruben\\Desktop\\Investigation\\updated_dataset_07_3_2018\\samples.csv', usecols=['id', 'device_id', 'timestamp', 'battery_state', 'battery_level'], parse_dates=['timestamp'])

samples_data.index = samples_data['id']
samples_data.drop('id', axis=1, inplace=True)

# devices_data.head()
samples_data.head()
```

```python
#basic stats
print('Total Devices =', len(devices_data))
num_dev_no_sampples = len(devices_data) - len(samples_data.groupby("device_id").count().index)
print('Devices with no samples =', num_dev_no_sampples, '(Percentage = ', round((num_dev_no_sampples/len(devices_data))*100, 2), ')')
```

```python
#basic stats and not charging devices samples count
dev_with_samples_list = list(samples_data.groupby("device_id").count().index)

dev_not_charging = []
for devID in dev_with_samples_list:
    if 'Not charging' in samples_data[samples_data.device_id == devID].battery_state.unique():
        dev_not_charging.append(devID)
        
print('Total devices with samples =', len(dev_with_samples_list), '(Percentage = ', round((len(dev_with_samples_list)/len(devices_data))*100, 2), ')')
print('Devices without ''Not charging'' samples =', len(dev_with_samples_list) - len(dev_not_charging))
print('Devices with ''Not charging'' samples =', len(dev_not_charging), '(Percentage = ', round((len(dev_not_charging)/len(dev_with_samples_list))*100, 2), ')')
```

```python
#not charging samples count per device
samples_not_charging = samples_data[samples_data.battery_state == 'Not charging']

samples_not_charging.groupby("device_id").count().sort_values(by='timestamp', ascending=False)
```
