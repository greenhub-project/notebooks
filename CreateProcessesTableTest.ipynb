```python
import pandas as pd
import numpy as np

import time
start_time = time.time()

chunksize = 10 ** 7
process_table = pd.DataFrame(columns=['sample_id'])


for df in pd.read_csv(r'C:\Users\ruben\Desktop\Investigation\updated_dataset_07_3_2018\app_processes.csv', usecols=['sample_id', 'name', 'application_label'], chunksize=chunksize):
    
    for index, row in df.iterrows():
        if row['sample_id'] not in process_table['sample_id'].values:
            process_table.reset_index(inplace=True, drop=True)
            process_table.loc[-1] = [0 for col_name in process_table.columns]
            process_table.loc[-1, 'sample_id'] = row['sample_id']


        if row['application_label'] != r'\N':
            if row['application_label'] in process_table.columns:
                process_table.loc[-1, row['application_label']] = 1
            else:
                process_table[row['application_label']] = 0
                process_table.loc[-1, row['application_label']] = 1

        else:
            if row['name'] in process_table.columns:
                process_table.loc[-1, row['name']] = 1
            else:
                process_table[row['name']] = 0
                process_table.loc[-1, row['name']] = 1

        

process_table.reset_index(inplace=True, drop=True)
process_table.head()
print("--- %s seconds ---" % (time.time() - start_time))

# df = pd.read_csv(r'C:\Users\ruben\Desktop\Investigation\updated_dataset_07_3_2018\app_processes.csv', usecols=['sample_id', 'name', 'application_label'], nrows=1000000)

# df.head(50)
```
